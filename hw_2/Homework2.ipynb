```
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

def f1(x, y):
    return (x ** 2 + y ** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 20, cmap='binary')
ax.plot([0],[0],marker='o',label='p1')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```
def f2(x, y):
    return -(x** 2 + y** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 20, cmap='binary')
ax.plot([0],[0],marker='o',label='p2')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```
def f3(x, y):
    return (x** 2 - y** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)

ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 20, cmap='binary')
ax.plot([0],[0],marker='o',label='p3')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Problem 2

```
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
np.random.seed(42)

# generate m random pairs (x, y)

m = 100
x1s = 2 * np.random.rand(m, 1)
x2s = 2 * np.random.rand(m, 1)
ys = 4 + 3 * x1s + 3 * x2s + np.random.rand(m, 1)

ax = plt.axes(projection='3d')
ax.scatter3D(x1s, x2s, ys)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
```

```
# number of epochs
epochs = 20
# learning rate
lr = 0.01

# training set
train_x1s = x1s[:80]
train_x2s = x2s[:80]
train_ys = ys[:80]
# test set
test_x1s = x1s[80:]
test_x2s = x2s[80:]
test_ys = ys[80:]

# initial weight values w1 and w2 and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)

for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_x1s[i] + w2 * train_x2s[i] + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_x1s[i]
    grad_w2 = (y_pred - train_ys[i]) * train_x2s[i]
    grad_b = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b

test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_x1s[i] + w2 * test_x2s[i] + b - test_ys[i]) ** 2
test_loss /= 20

test_loss

pred_ys = w1 * test_x1s + w2 * test_x2s + b

ax = plt.axes(projection='3d')
ax.scatter3D(test_x1s, test_x2s, test_ys)
ax.scatter3D(test_x1s, test_x2s, pred_ys)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
```
